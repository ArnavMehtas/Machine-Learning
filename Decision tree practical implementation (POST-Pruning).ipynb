```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
#independent feature and dependent features
X=df.iloc[:,:-1]
y=iris.target
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Post pruning technique
treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(X_train , y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel , filled=True)
```

```python
# using hyper parameter:

treemodel2 = DecisionTreeClassifier(max_depth=2) #postproning by max_depth
treemodel2.fit(X_train , y_train)
```

```python
plt.figure(figsize=(5,4))
tree.plot_tree(treemodel2,filled=True)
```

```python
y_pred=treemodel2.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred,y_test))
```
